```python
## Original

all_tweets = []

for twitter_username in list_reporters:
    try: 
        print(f"Getting tweets from {twitter_username}")

        user_tweets = []

        new_tweets = api.user_timeline(screen_name = twitter_username, count = 200, exclude_replies = True, include_rts = False)
        
        user_tweets.extend(new_tweets)
        oldest = user_tweets[-1].id - 1
        # oldest_date = datetime.fromtimestamp(oldest)

        while len((new_tweets)) > 0:
            print(f"getting tweets before {oldest}")
            
            #Use oldest to prevent duplicates
            new_tweets = api.user_timeline(screen_name = twitter_username, count = 200, exclude_replies = True, include_rts = False, max_id = oldest)
            
            #Save most recent tweets
            user_tweets.extend(new_tweets)
            
            #Update ID of oldest
            oldest = user_tweets[-1].id - 1
            
            print(f"...{len(user_tweets)} tweets downloaded so far")

        all_tweets.extend(user_tweets)
    except Exception as e:
        print(e)
        print(f"Error getting tweets from {twitter_username}")
```

```python
df = pd.DataFrame(map(lambda x: x._json, all_tweets))
```
